```python
#NORMALİZASYON ÇEŞİTLERİ
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Veri setini yükleyin
veri_seti = pd.read_csv("dataset2.csv")  # veri_seti.csv dosyanızın adını ve yolunu vermelisiniz

# Min-max normalizasyonu için scaler'ı tanımlayın
scaler = MinMaxScaler()

# Normalize edilecek sütunları seçin (Class variable (0 or 1) sütunu hariç)
normalize_edilecek_sutunlar = ["Number of times pregnant",
                               "Plasma glucose concentration a 2 hours in an oral glucose tolerance test",
                               "Diastolic blood pressure (mm Hg)",
                               "Triceps skinfold thickness (mm)",
                               "2-Hour serum insulin (mu U/ml)",
                               "Body mass index (weight in kg/(height in m)^2)",
                               "Diabetes pedigree function",
                               "Age (years)"]

# Seçilen sütunları normalize edin
veri_seti[normalize_edilecek_sutunlar] = scaler.fit_transform(veri_seti[normalize_edilecek_sutunlar])

# Normalize edilmiş veri setini görüntüleyin
print(veri_seti.head())
```

```python
import numpy as np
import pandas as pd

data = pd.read_csv("dataset2.csv")

# Ortalama ve standart sapmayı buluyoruz
mean = data.mean()
std = data.std()

# Z-Score Normalizasyonu
normalized_data = (data - mean) / std

print("\nNormalleştirilmiş Veriler:")
print(normalized_data)
```

```python
import numpy as np
import pandas as pd

def sigmoid_normalize(data):

  return 1 / (1 + np.exp(-data))

data = pd.read_csv("dataset2.csv")
normalized_data = sigmoid_normalize(data.copy())

print("\nNormalleştirilmiş Veriler (Sigmoid):")
print(normalized_data)
```

```python
import numpy as np
import pandas as pd

def median_normalization(data):
  """
  Verileri ortalama olarak medyan ve ölçek faktörü olarak aralık kullanarak normalleştirir.
  """
  median_val = data.median()
  range_val = data.max() - data.min()
  normalized_data = (data - median_val) / range_val
  return normalized_data

data = pd.read_csv("dataset2.csv")
normalized_data = median_normalization(data.copy())  # Operate on a copy

print("\nNormalleştirilmiş Veriler (Medyan ve Aralık):")
print(normalized_data)
```
