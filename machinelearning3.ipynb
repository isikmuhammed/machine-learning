```python
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree

dataset = pd.read_csv('dataset2.csv')

dataset.columns = ['pregnant', 'glucose', 'blood pressure', 'skin thickness', 'insulin', 'bmi', 'diabetes pedigree', 'age', 'class']

X = dataset.drop('class', axis=1)
y = dataset['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression

# Çoklu Doğrusal Regresyon modelini oluşturma
linear_reg_model = LinearRegression()

# Modeli eğitme
linear_reg_model.fit(X_train, y_train)

# Eğitim verisi üzerinde tahmin yapma
y_train_pred_linear = linear_reg_model.predict(X_train)

# Test verisi üzerinde tahmin yapma
y_test_pred_linear = linear_reg_model.predict(X_test)
# Çoklu Doğrusal Regresyon için tahminlerin yazdırılması
print("Çoklu Doğrusal Regresyon - Eğitim verisi tahminleri:")
print(y_train_pred_linear)
print("\nÇoklu Doğrusal Regresyon - Test verisi tahminleri:")
print(y_test_pred_linear)
print("\n")
```

```python
from sklearn.linear_model import LogisticRegression

# Multinominal Lojistik Regresyon modelini oluşturma
logistic_reg_model = LogisticRegression(multi_class='multinomial', solver='lbfgs')

# Modeli eğitme
logistic_reg_model.fit(X_train, y_train)

# Eğitim verisi üzerinde tahmin yapma
y_train_pred_logistic = logistic_reg_model.predict(X_train)

# Test verisi üzerinde tahmin yapma
y_test_pred_logistic = logistic_reg_model.predict(X_test)
# Multinominal Lojistik Regresyon için tahminlerin yazdırılması
print("Multinominal Lojistik Regresyon - Eğitim verisi tahminleri:")
print(y_train_pred_logistic)
print("\nMultinominal Lojistik Regresyon - Test verisi tahminleri:")
print(y_test_pred_logistic)
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score

# Çoklu Doğrusal Regresyon performans metrikleri
mse_linear = mean_squared_error(y_test, y_test_pred_linear)

# Multinominal Lojistik Regresyon performans metrikleri
accuracy_logistic = accuracy_score(y_test, y_test_pred_logistic)

print("Çoklu Doğrusal Regresyon - Test verisi MSE:", mse_linear)
print("Multinominal Lojistik Regresyon - Test verisi accuracy:", accuracy_logistic)



```
